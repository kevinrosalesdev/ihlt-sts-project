```python
import nltk
from nltk.wsd import lesk

from nltk import pos_tag
from scipy.stats import pearsonr
from nltk.metrics import jaccard_distance
```

## Senses

```python
# Define Lesk
from nltk.stem import WordNetLemmatizer

def leskk(c,p):
    if p[1][0] in {'N','V'}:
        return lesk(c,p[0].lower(), pos=p[1][0].lower())
    return p[0]
```

```python
# Get the plain text
file = open("STS.input.txt")
# define the placeholders
setence_ids = []
sentence_tokenized = {}
sentences_list=[]
# Loop to extract each id with sentences
for i in file:
    id = i[0:3]
    setence_ids.append(id)
    sentences = nltk.sent_tokenize(i[4:])
    # Get plain sentence in a list
    sentences_list.append(sentences)
    # Get tokenized sentence
    sentence_tokenized[id] = [nltk.word_tokenize(s) for s in sentences]
## Add sentences
sentence3_2 = sentence_tokenized.get('id3')[1]
sentence3_3 = sentence_tokenized.get('id3')[2]
sentence3_2.extend(sentence3_3)
# Change the sentence value of the id3
sentence_tokenized['id3'][1] = sentence3_2

print (sentences_list)
```

```python
## Lesk pairs
post_tag_pairs = {}
pairs_lesk = {}
count=0
for id in setence_ids:
    post_tag_pairs[id] = pos_tag(sentence_tokenized.get(id)[0]),pos_tag(sentence_tokenized.get(id)[1])
#     print(set(post_tag_pairs.get(id)[0]))
#     print(set(post_tag_pairs.get(id)[1]))
    #get senses in pairs
    print(sentences_list[count][0])
    print(sentences_list[count][1])
    print("\n")
    pairs_lesk[id]=[leskk(sentences_list[count][0],pair) for pair in post_tag_pairs[id][0]],[leskk(sentences_list[count][1],pair) for pair in post_tag_pairs[id][1]]
    count=count+1
```

```python
# jaccard distances considering senses
jd_senses = []
for id in setence_ids:
    print(set(pairs_lesk.get(id)[0]))
    print(set(pairs_lesk.get(id)[1]))
    print("\n")
    distance = jaccard_distance(set(pairs_lesk.get(id)[0]),set(pairs_lesk.get(id)[1]))
    jd_senses.append(distance)

print("\nJaccard distance considering senses: %s"%jd_senses)
```

## Session 2: Document

```python
# jaccard distances for sentences tokenized
jd_words = []
for id in setence_ids:
#     print(set(sentence_tokenized.get(id)[0]))
#     print(set(sentence_tokenized.get(id)[1]))
#     print("\n")
    distance = jaccard_distance(set(sentence_tokenized.get(id)[0]),
                           set(sentence_tokenized.get(id)[1]))
    jd_words.append(distance)
print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words)
```

## Session 3: Morphology

```python
# Define Wornet Lematizer
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

def lemmatize(p):
    if p[1][0] in {'N','V'}:
        return wnl.lemmatize(p[0].lower(), pos=p[1][0].lower())
    return p[0]
```

```python
## lemmas
post_tag_pairs = {}
pairs_lemmas = {}

for id in setence_ids:
    #capturing the part of speech tag for each pair of sentences
    post_tag_pairs[id] = pos_tag(sentence_tokenized.get(id)[0]),pos_tag(sentence_tokenized.get(id)[1])
    #get lemmas in pairs
    pairs_lemmas[id]=[lemmatize(pair) for pair in post_tag_pairs[id][0]],[lemmatize(pair) for pair in post_tag_pairs[id][1]]
# print(pairs_lemmas)
```

```python
# jaccard distances for lemmas
jd_lemmas = []
for id in setence_ids:
    #print(set(pairs_lemmas.get(id)[0]))
    #print(set(pairs_lemmas.get(id)[1]))
    distance = jaccard_distance(set(pairs_lemmas.get(id)[0]),set(pairs_lemmas.get(id)[1]))
    jd_lemmas.append(distance)

print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas)
```

## Result

### Pearson correlation & Jaccard distances

```python
# We could get the plain text with an extraction again: gs_file = open("STS.gs.txt")  or we could simply create the list with the trivial values as it is not a complex file:
# Gold standard
gs = [0,1,2,3,4,5]

pearson_words = pearsonr(gs, jd_words)
pearson_lemmas = pearsonr(gs, jd_lemmas)
pearson_senses= pearsonr(gs, jd_senses)
# Jaccard distances
print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words)
print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas)
print("\nJaccard distance for pairs of lemmas: %s"%jd_senses)
# Global value
print('\nPerson correlation value for words: %f'%(pearson_words[0]))
print('Person correlation value for lemmas: %f'%(pearson_lemmas[0]))
print('Person correlation value for senses: %f'%(pearson_senses[0]))
```

### Conclusion

As it is, a better performance is reached using the sense disambiguation. However, given the fact that only in the first pair gave a better performance than the lemmas. It's worth trying with a higher dataset, as shown bellow. And after the process is evident that senses is far superior, and this is reasonable because it's a dissambiguation given the context of the sentence.
