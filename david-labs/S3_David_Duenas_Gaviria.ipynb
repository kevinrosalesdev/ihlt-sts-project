```python
import nltk
from scipy.stats import pearsonr
from nltk.metrics import jaccard_distance
from nltk import pos_tag
```

```python
# Define Wornet Lematizer
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

def lemmatize(p):
    if p[1][0] in {'N','V'}:
        return wnl.lemmatize(p[0].lower(), pos=p[1][0].lower())
    return p[0]
```

#### Activate only to download
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

```python
# Get the plain text
file = open("STS.input.txt")
# define the placeholders
setence_ids = []
sentence_tokenized = {}
# Loop to extract each id with sentences
for i in file:
    id = i[0:3]
    setence_ids.append(id)
    sentences = nltk.sent_tokenize(i[4:])
    sentence_tokenized[id] = [nltk.word_tokenize(s) for s in sentences]
print (sentence_tokenized)

## add sentences
sentence3_2 = sentence_tokenized.get('id3')[1]
sentence3_3 = sentence_tokenized.get('id3')[2]
sentence3_2.extend(sentence3_3)
# Change the sentence value of the id3
sentence_tokenized['id3'][1] = sentence3_2
print(type(sentence_tokenized))
print(type(setence_ids))
```

```python
# jaccard distances for sentences tokenized
jd_words = []
for id in setence_ids:
    print(set(sentence_tokenized.get(id)[0]))
    print(set(sentence_tokenized.get(id)[1]))
    print("\n")
    distance = jaccard_distance(set(sentence_tokenized.get(id)[0]),
                           set(sentence_tokenized.get(id)[1]))
    jd_words.append(distance)
print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words)
```

## Lemmas

```python
## lemmas
post_tag_pairs = {}
pairs_lemmas = {}

for id in setence_ids:
#     print(set(sentence_tokenized.get(id)[0]))
#     print(set(sentence_tokenized.get(id)[1]))
    #capturing the part of speech tag for each pair of sentences
    post_tag_pairs[id] = pos_tag(sentence_tokenized.get(id)[0]),pos_tag(sentence_tokenized.get(id)[1])
    print(set(post_tag_pairs.get(id)[0]))
    print(set(post_tag_pairs.get(id)[1]))
    #get lemmas in pairs
    pairs_lemmas[id]=[lemmatize(pair) for pair in post_tag_pairs[id][0]],[lemmatize(pair) for pair in post_tag_pairs[id][1]]
print(pairs_lemmas)
```

```python
# jaccard distances for lemmas
jd_lemmas = []
for id in setence_ids:
    #print(set(pairs_lemmas.get(id)[0]))
    #print(set(pairs_lemmas.get(id)[1]))
    distance = jaccard_distance(set(pairs_lemmas.get(id)[0]),set(pairs_lemmas.get(id)[1]))
    jd_lemmas.append(distance)

print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas)
```

## Results

```python
# Showing the two jaccard distances
print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words)
print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas)
```

All the jaccard distances were the same except for the second pair of sentences: that computed a slightly shorter jaccard distance of 0.66 compared to 0.73 in the previous lab. This is due to the verb 'invade' as in the second sentence it's in past tense and converted to his lemma allowes a lower jaccard distance.

```python
# We could get the plain text with an extraction again: gs_file = open("STS.gs.txt")  or we could simply create the list with the trivial values as it is not a complex file:
# Gold standard
gs = [0,1,2,3,4,5]

pearson_words = pearsonr(gs, jd_words)
pearson_lemmas = pearsonr(gs, jd_lemmas)
# Jaccard distances
print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words)
print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas)
# Global value
print('\nPerson correlation value for words: %f'%(pearson_words[0]))
print('Person correlation value for lemmas: %f'%(pearson_lemmas[0]))
```

Then as it is expected, the person correlation value performed slightly better compared to the second session exercise.

### Trying with test data

```python
# Get the plain text
file = open("msr_paraphrase_test.txt")
# define the placeholders
test_setence_ids = []
test_sentence_tokenized = {}
# Loop to extract each id with sentences
for i in file:
    id = i[0:18]
    test_setence_ids.append(id)
    test_sentences = nltk.sent_tokenize(i[18:])
    test_sentence_tokenized[id] = [nltk.word_tokenize(s) for s in test_sentences]

# Cleaning the data (deleting the first sentence with the description of the document)
del test_sentence_tokenized['\ufeffQuality\t#1 ID\t#2 ']
test_setence_ids.remove(test_setence_ids[0])
#print (test_setence_ids)

print(len(test_sentence_tokenized))
print(len(test_setence_ids))
```

```python
# jaccard distances for sentences tokenized
jd_words = []
#print(set(test_sentence_tokenized.get(id)[2]))
for id in test_setence_ids[:500]:
    #print(set(test_sentence_tokenized.get(id)[0]))
    #print('\r')
    #print(set(test_sentence_tokenized.get(id)[1]))
    #print('\n\n\n')
    distance = jaccard_distance(set(test_sentence_tokenized.get(id)[0]),
                           set(test_sentence_tokenized.get(id)[1]))
    jd_words.append(distance)

print("\nJaccard distance for pairs of sentences tokenized: %s"%jd_words[:50])
```

```python
## lemmas
post_tag_pairs = {}
pairs_lemmas = {}

for id in test_setence_ids[:500]:
    #print(set(sentence_tokenized.get(id)[0]))
    #print(set(sentence_tokenized.get(id)[1]))
    #capturing the part of speech tag for each pair of sentences
    post_tag_pairs[id] = pos_tag(test_sentence_tokenized.get(id)[0]),pos_tag(test_sentence_tokenized.get(id)[1])
    #get lemmas in pairs
    pairs_lemmas[id]=[lemmatize(pair) for pair in post_tag_pairs[id][0]],[lemmatize(pair) for pair in post_tag_pairs[id][1]]
print(pairs_lemmas)
```

```python
# jaccard distances for lemmas
jd_lemmas = []
for id in test_setence_ids[:500]:
    #print(set(pairs_lemmas.get(id)[0]))
    #print(set(pairs_lemmas.get(id)[1]))
    distance = jaccard_distance(set(pairs_lemmas.get(id)[0]),set(pairs_lemmas.get(id)[1]))
    jd_lemmas.append(distance)

print("\nJaccard distance for pairs of lemmas: %s"%jd_lemmas[:50])
```

Giving a quick look to the jaccard distances of the 500 pair of sentences, we'll have a better understanding that the lemmas indeed help to compute a lower jaccard distance.

```python
# We could get the plain text with an extraction again: gs_file = open("STS.gs.txt")  or we could simply create the list with the trivial values as it is not a complex file:
# Gold standard
gs = [0,1,2,3,4,5]

pearson_words = pearsonr(gs, jd_words[20:26])
pearson_lemmas = pearsonr(gs, jd_lemmas[20:26])
# Global value
print('\nPerson correlation value for words: %f'%(pearson_words[0]))
print('Person correlation value for lemmas: %f'%(pearson_lemmas[0]))
```

The performance of aleatory pair of sentences was better with lemmas.
